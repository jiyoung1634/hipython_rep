# melt()

- id_vars = 유지할 변수 리스트(튜플, 행렬 등)
- value_vars : 피봇 되돌리기 값으로 쓰일 열 목록
- var_name : 만들어지는 새 컬럼의 이름 (예 : variable)
- value_name : 새 컬럼의 값의 컬럼 이름 (예 : value)

```python
#직원 명단 DataFrame 생성
data = {
    'Name':['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age':[25, 45, 35, 32, 28],
    'salary':[5000, 7000, 6500, 5500, 6000]
    }
df = pd.DataFrame(data)
df
```

```python
#연령대별로 만들어주는 사용자 정의 함수
def Age_category(age):
    if age>=20 and age<30:
        return '20대'
    elif age>=30 and age<40:
        return '30대'
    elif age>=40 and age<50:
        return '40대'
```

```python
df['Age'].apply(Age_category)
```

```python
# 강사님 답
def age_category(age):
    if age<30:
        return '20대'
    elif age<40:
        return '30대'
    else:
        return '40대이상'

age_category(26)
```

```python
#Category 컬럼으로 추가
df['Age_Category'] = df['Age'].apply(age_category)
df
```

```python
#보너스 30%
def calc_bonus(salary):
    return salary*0.30
df['Bonus'] = df['salary'].apply(calc_bonus)
df
```

```python
#보너스 람다로 바꾸기
df['Bonus_lambda'] = df['salary'].apply(lambda x : x*0.3)
df
```

# 4. 그룹 연산하기

groupby(), aggregation, 집계

```python
df = pd.read_csv('./data/gapminder.tsv', sep='\t')
df
```

```python
df['country'].unique().shape   #unique함수로 한개씩만 꺼내볼 수 있다
```

```python
df.groupby('country')['lifeExp'].mean()   #그룹하려는 country의 lifeExp
```

count(), size(), mean(), std()
quantile()

## numpy함수도 가능
np.max(), np.sum(), np.var()

```python
df.groupby('continent')['lifeExp'].count()
```

```python
import numpy as np
#여러개를 적용해서 합할 때 agg()
agg_df = df.groupby('continent')['lifeExp'].agg([np.mean, np.std, np.min, np.max])   #aggregation에 적용하려는 함수 넣기(여러개는 리스트로 넣기)
agg_df
```

위는 하나의 변수(continent)에 여러 함수들을 적용한거고 반대로 <br>
여러 변수에도 적용할 수 있다(딕셔너리로 변수의 이름과 적용할 함수의 이름)

```python
agg_df = df.groupby('continent').agg({'lifeExp':'mean',
                                      'pop':'median',
                                      'gdpPercap':'median'})
agg_df
```
