```python
store_name = '또!젤라또'

menu_list = [
    {'name':'또젤라또','count':'1가지맛', 'price':3000},
    {'name':'또또젤라또','count':'2가지맛', 'price':5000},
    {'name':'또또또젤라또','count':'3가지맛', 'price':7000}
]

flavor_list = [
    {'flavor':'달콤초코봄바람'},
    {'flavor':'새콤달콤딸기정원'},
    {'flavor':'요거트빛파도'},
    {'flavor':'말차소풍'},
    {'flavor':'체리팝송'},
    {'flavor':'포도에몽'},
    {'flavor':'바닐라하모니'},
    {'flavor':'쿠키크러쉬'},
    {'flavor':'소금바다우유'},
    {'flavor':'블루베리썸머'},
    {'flavor':'스윗피치'},
    {'flavor':'레몬샤워'},
]

while True:
    orders = []

    while True:
        # 메뉴 선택
        print("="*45)
        print(f'{store_name:^40}')
        print("="*45)
        print("어서오세요 아이스크림을 또 주는 또젤라또입니다\n")
    
        for i, menu in enumerate(menu_list):
            print(f'{i + 1}. {menu["name"]:<6} {menu["count"]:>12} {menu["price"]:>10,}원')
        print("="*45)
    
        order_input = input('주문할 메뉴 번호를 입력해주세요 (또는 Q 입력 시 종료): ').strip().lower()
        if order_input == 'q':
            break
        if not order_input.isdigit() or not (1 <= int(order_input) <= len(menu_list)):
            print("잘못된 입력입니다. 다시 시도해주세요.\n")
            continue
    
        order_num = int(order_input) - 1
        selected_menu = menu_list[order_num]
        flavor_count = int(selected_menu['count'][0])

        # 맛 선택
        while True:
            print(f"\n'{selected_menu['name']}' 선택 — {flavor_count}가지 맛을 골라주세요")
            for i, fla in enumerate(flavor_list):
                print(f"{i+1}. {fla['flavor']}")
    
            flavor_input = input("\n맛 번호들을 입력하세요 (예: 1, 3, 5) / 0 입력시 메뉴 선택으로 돌아가기: ").strip()
            if flavor_input == '0':
                print("맛 선택을 취소하고 메뉴로 돌아갑니다.\n")
                break
    
            parts = flavor_input.split(',')
            flavor_nums = [int(p.strip()) - 1 for p in parts if p.strip().isdigit()]
    
            if len(flavor_nums) != flavor_count or any(i < 0 or i >= len(flavor_list) for i in flavor_nums):
                print(f"정확히 {flavor_count}가지 맛을 정확히 입력해주세요.")
                continue
    
            selected_flavors = [flavor_list[i]['flavor'] for i in flavor_nums]

            # 주문 확인
            print("\n선택한 주문:")
            print(f"메뉴: {selected_menu['name']}")
            for idx, name in enumerate(selected_flavors):
                print(f" - 맛 {idx+1}: {name}")
            print(f"가격: {selected_menu['price']:,}원")
    
            confirm = input("\n이 주문을 장바구니에 추가할까요? (Y: 추가 / N: 취소): ").strip().lower()
            if confirm == 'y':
                orders.append({
                    'menu_name': selected_menu['name'],
                    'flavor_names': selected_flavors,
                    'price': selected_menu['price']
                })
                print("주문이 장바구니에 추가되었습니다.\n")
            else:
                print("주문이 취소되었습니다.\n")
    
            break  # 맛 선택 루프 탈출 → 다음으로 진행
    
        another = input("추가 주문하시겠습니까? (Y/N): ").strip().lower()
        if another != 'y':
            break

    # 주문 내역 확인 및 취소
    while True:
        print("\n전체 주문 내역:")
        if not orders:
            print("현재 주문이 없습니다.")
            break
    
        for i, order in enumerate(orders):
            print(f"{i+1}. {order['menu_name']} — {order['price']:,}원")
            for idx, name in enumerate(order['flavor_names']):
                print(f"   - 맛 {idx+1}: {name}")
    
        cancel_q = input("\n주문 내역 중 취소할 주문이 있습니까? (Y/N): ").strip().lower()
        if cancel_q == 'n':
            break
        elif cancel_q != 'y':
            print('!(Y/N)로 입력해주세요!')
            continue
    
        del_input = input("취소할 주문 번호를 입력하세요 (0 입력시 완료): ").strip()
        if del_input == '0':
            break
        if del_input.isdigit() and 1 <= int(del_input) <= len(orders):
            del_index = int(del_input) - 1
            removed = orders.pop(del_index)
            print(f"'{removed['menu_name']}' 주문이 삭제되었습니다.")
        else:
            print("잘못된 번호입니다.")

    # 최종 결제
    total_price = sum(order['price'] for order in orders)
    print('\n' + '='*45)
    print(f"총 결제 금액: {total_price:,}원")

    if total_price == 0:
        print("남은 주문이 없습니다. 처음으로 돌아갑니다.\n")
        continue

    point_q = input('포인트 적립하시겠습니까? (Y/N): ').strip().lower()
    if point_q == 'y':
        membership = input("전화번호를 입력해주세요: ")
        saved_point = total_price // 100
        existing_point = 250
        total_point = existing_point + saved_point
        print(f"{saved_point}포인트 적립 완료 → 총 {total_point}포인트")
    else:
        membership = "고객"

    print(f"\n{membership}님, 주문 감사합니다! 또 오세요.")
    print("="*45)
```

```python
import time


# # 메뉴 저장 리스트 -> 메뉴 추가 시 여기서.
menu = [
    {'name':'샤프', 'price': 6000},
    {'name':'볼펜', 'price': 3000},
    {'name':'지우개', 'price': 1000},
    {'name':'화이트', 'price': 3000},
    {'name':'공책', 'price': 3500}
]

points = {}

while True:

    space = ' '*7
    print('╔' + '═' * 40 + '╗')
    print(f'\t{space}빵꾸똥꾸 문구야')
    print(f'{"에 오신 것을 환영합니다!":>25}')
    print('╚' + '═' * 40 + '╝')


    print(f'{"번호"}\t\t{"제품명"}\t\t{"가격"}')
    print('-'*44)

    for i, m in enumerate(menu):
        print(f'{i+1}\t\t{m["name"]}\t\t{m["price"]:,}원')
    print('-'*44)

    # 다중 주문 저장 리스트
    order_list = []


    while True:

        try: ### 제품번호 범위 및 숫자 외 입력 방지 ###
            ord_num = int(input('주문하실 제품 번호를 선택하세요: ')) -1

            if not (0 <= ord_num < len(menu)): 
                print('='*44)
                print(f'{"[알림] 메뉴에 없는 제품 번호입니다.":^35}')
                print(f'{"[알림] 목록에 있는 제품 번호만 입력해 주세요.":^35}')
                print('='*44)
                continue 


        except ValueError:  
            print('='*44)
            print(f'{"[알림] 잘못된 입력입니다.":^35}')
            print(f'{"[알림] 제품 번호를 숫자로 입력해 주세요.":^35}')
            print('='*44)
            continue 
        
        menut = menu[ord_num]['name']

        print('='*44)
        print(f'{"고객님이 선택하신 제품은":^33}')
        print(f'{menut + " 입니다.":^33}')
        print(f'{"몇 개를 드릴까요?":^33}')
        
        go_to_start = False
        
        while True: ### 수량값 음수 및 문자 입력 방지 ###
            try:

                qty = input('\t수량을 입력하세요(취소: x): ')

                if qty == 'x':
                    go_to_start = True
                    break
                else:
                    qty = int(qty)
                
                if qty <=0:
                    print('='*44)
                    print(f'{"[알림]수량은 1개 이상 입력해주세요.":^35}')
                    print('='*44)
                    continue
                break

            except ValueError:
                print('='*44)
                print(f'{"[알림]다시 입력해주세요.":^35}')
                print('='*44)

        if go_to_start:
            break
                
        # 만약 같은 걸 더 주문하면 order_list 에 추가 말고 qty 만 +++
        flag = True
        for i in range(len(order_list)):
            if order_list[i]['name'] == menu[ord_num]["name"]:
                order_list[i]['qty'] += qty
                flag = False
                break

        # 새로운 거 주문 하면 append
        if flag:
            order_list.append({
                'name': menut,
                'qty': qty,
                'price': menu[ord_num]['price']
            })

        a = input('더 주문하시겠습니까? (y/n): ')
            # y/n 입력 유효성 검사
        while a.lower() not in ['y', 'n']: ### y,n 외 입력 방지 ###
            print('='*44)
            print(f'{"[알림] 잘못된 입력입니다.":^35}')
            print(f'{"[알림] 'y' 또는 'n'으로만 입력해 주세요.":^35}')
            print('='*44)
            a = input('더 주문하시겠습니까? (y/n): ')
            
        if a.lower() == 'n':
            break

        print('-'*44)

    if not go_to_start:
        # 총액 계산 및 주문 출력
        total = 0
    
        print('='*44)
        print(f'\t\t{"주문 내역"}')
        print('-'*44)
    
        for order in order_list:
            item_total = order['qty'] * order['price']
            total += item_total
    
            print(f'{order["name"]}\t{order["qty"]}개\t-\t{item_total:,}원')
    
        print('-'*44)
        print(f'총 결제 금액: {total:,}원')
        print('='*44)
    
        # 포인트 적립 여부 y/n 유효성 검사
        ab = input('포인트를 적립하시겠습니까? (y/n): ')
        
        while ab.lower() not in ['y', 'n']:
            print('=' * 44)
            print(f'{"[알림] 잘못된 입력입니다.":^35}')
            print(f'{"[알림] 'y' 또는 'n'으로만 입력해 주세요.":^35}')
            print('=' * 44)
            ab = input('포인트를 적립하시겠습니까? (y/n): ')
    
        # 'y'일 경우 포인트 적립
        if ab.lower() == 'y':
            phone = input('전화번호를 입력해 주세요: ')
            
            # 전화번호가 숫자가 아닐 경우 재입력 요구
            while not phone.isdigit() or len(phone) < 11 or phone == "x":
                print('=' * 44)
                print(f'{"[알림] 잘못된 입력입니다.":^35}')
                print(f'{"[알림] 전화번호는 숫자만 입력해 주세요.":^35}')
                print('=' * 44)
                phone = input('전화번호를 다시 입력해 주세요: ')
                
        if ab.lower() == 'y':
            user_id = f'user_{phone[-4:]}'
            point = int(total * 0.05)
        
            if user_id in points :
                points[user_id] += point
            else :
                points[user_id] = point
    
            print('=' * 44)
            print(f'{user_id} 고객님의 적립포인트는 {points[user_id]:,}점 입니다.')


        print('='*44)
        print(f'{"결제가 완료되었습니다.":^37}')
        print(f'{"잠시만 기다려주세요.":^37}')

    print('='*44)
    print(f'{"감사합니다.":^40}')
    print(f'{"3초 뒤에 시작화면으로 돌아갑니다.":^37}')
    time.sleep(1)
    print('='*44)
    print(f'{"2초 뒤에 시작화면으로 돌아갑니다.":^37}')
    time.sleep(1)
    print('='*44)
    print(f'{"1초 뒤에 시작화면으로 돌아갑니다.":^37}')
    time.sleep(1)
```

```python
import time
  #################메뉴리스트정의
menu_list = [ 
    {'name': '요거트 아이스크림', 'price': 3000},
    {'name': '바나나', 'price': 4000},
    {'name': '딸기', 'price': 5000},
    {'name': '벌집꿀', 'price': 7000},
    {'name': '그레놀라', 'price': 7000},
    {'name': '초코쉘', 'price': 7000},
]
final_break = False
while True:
    ######################################초기화면 (메뉴판)################################################
    print('*'*40)
    print('어서오세요, ‘카페’에 오신것을 환영합니다:)\n')
    
    
    print("Menu\t\t\tPrice")
    print("-" * 30)
    print("요거트 아이스크림(250g)\t3,000")
    print("-" * 30)
    
    print("\nNo.  Topping\t\tPrice")
    print("-" * 30)
    for idx, item in enumerate(menu_list, start=1):
        print(f"{idx:<5} {item['name']:<10}\t{item['price']:,}")
    print("-" * 30)
    print("*" * 40)
    
    ##############카트 초기화, 기본 상품(요거트 아이스크림) 장바구니에 자동추가
    cart = [] 
    default_item = menu_list[0]
    cart.append({'name': default_item['name'], 'price': default_item['price'], 'quantity': 1})
    total_price = 3000
    cancle = False
    ##############[2]장바구니 관리 루프 시작
    while True:
       #########장바구니 출력
        print("\n" + "=" * 40)
        print("현재 장바구니")
        for i, item in enumerate(cart, start=1):
            print(f"{i}. {item['name']} x {item['quantity']} = {item['price'] * item['quantity']:,}원")
        print("-" * 40)
        print(f"Total = {total_price:,}원")
      ########################메뉴선택안내
       ##########################1.결제화면 2. 삭제 메뉴화면 3. 추가 메뉴화면 4. 초기화면
        print("메뉴를 확정하려면 1")
        print("메뉴를 삭제하려면 2")
        print("메뉴를 추가하려면 3")
        print("메뉴를 취소하려면 4")
        ############################# 99번 입력 시 무한루프 끝 #####################################
        select = int(input("번호를 입력해주세요 : "))
        ####################[2-1]주문 확정
        if select == 1:
            break
        ####################[2-2]메뉴 삭제
        elif select == 2:
           #################삭제할 메뉴 번호 입력받고 수량 삭제 처리
            print("\n" + "=" * 40)
            print("현재 장바구니")
            for i, item in enumerate(cart, start=1):
                print(f"{i}. {item['name']} x {item['quantity']} = {item['price'] * item['quantity']:,}원")
            print("-" * 40)
            print(f"Total = {total_price:,}원")
        
            del_no = int(input("삭제할 메뉴의 번호를 골라주세요 : ")) - 1
            if 0 <= del_no < len(cart):
                item = cart[del_no]
                print(f"'{item['name']}'의 현재 수량: {item['quantity']}")
                del_qty = int(input("몇 개를 삭제하시겠습니까? : "))
        
                if del_qty >= item['quantity'] or del_qty <= 0:
                    # 전체 삭제
                    total_price -= item["price"] * item["quantity"]
                    cart.pop(del_no)
                    print(f"'{item['name']}' 전체가 장바구니에서 삭제되었습니다.")
                else:
                    # 일부 수량만 삭제
                    item['quantity'] -= del_qty
                    total_price -= item["price"] * del_qty
                    print(f"'{item['name']}' {del_qty}개가 삭제되었습니다.")
        
            else:
                print("올바른 번호를 입력하세요.")
        
            print("\n" + "=" * 40)
            print("현재 장바구니")
            for i, item in enumerate(cart, start=1):
                print(f"{i}. {item['name']} x {item['quantity']} = {item['price'] * item['quantity']:,}원")
            print("-" * 40)
            print(f"Total = {total_price:,}원")
    ############################# [2-3] 메뉴 추가
        elif select == 3:
     ######################## 메뉴 리스트 보여주고 선택한 토핑 추가
            print("\n메뉴 목록:")
            for i, item in enumerate(menu_list, start=1):
                print(f"{i}. {item['name']} : {item['price']:,}원")
            add_topping = int(input('추가할 토핑 번호를 골라주세요 :')) - 1
            qty = int(input('수량을 선택해주세요 : '))
    
            selected_item = menu_list[add_topping]
            found = False
    
            for item in cart:
                if item['name'] == selected_item['name']:
                    item['quantity'] += qty
                    found = True
                    break
    
            if not found:
                cart.append({'name': selected_item['name'], 'quantity': qty, 'price': selected_item['price']})
    
            total_price += selected_item['price'] * qty
    
            print("\n" + "=" * 40)
            print("현재 장바구니")
            for i, item in enumerate(cart, start=1):
                print(f"{i}. {item['name']} x {item['quantity']} = {item['price'] * item['quantity']:,}원")
            print("-" * 40)
            print(f"Total = {total_price:,}원")
       ###############[2-4]주문 취소
        elif select == 4:
            print("주문이 취소됩니다.")
            cancle = True
            break
        ################# [2-5] 프로그램 종료 명령어
        elif select == 99:
            final_break = True
            break
    ################[3] 외부 루프 종료 조건
    if final_break:
        break
   
    ################# [4] 주문 취소 시 초기화면으로 복귀
    if cancle:
        print("초기화면으로 돌아갑니다.")
        for i in range(5, 0, -1):
            print(f"{i}   ", end='\r', flush=True)
            time.sleep(1)
        print("   ", end='\r' )
        continue
   ###################### [5] 결제 여부 확인
    pay = input("주문이 확정되었습니다. 결제를 하시겠습니까? (y/n)")
    if pay == 'y':
        print(f"{total_price:,}원 결제하겠습니다. 카드를 삽입해주세요.")
        print("결제가 완료되었습니다.")
        print("초기화면으로 돌아갑니다.")
        for i in range(5, 0, -1):
            print(f"{i}   ", end='\r', flush=True)
            time.sleep(1)
        print("   ", end='\r' )
        continue
    else:
        print("초기화면으로 돌아갑니다.")
        for i in range(5, 0, -1):
            print(f"{i}   ", end='\r', flush=True)
            time.sleep(1)
        print("   ", end='\r' )
        continue
```
